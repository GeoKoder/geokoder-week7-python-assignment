```python
# Imports the pandas and matplotlib library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Imports the titanic dataset
titanic = pd.read_csv('Titanic-Dataset.csv')
```

```python
# Display the first few rows
titanic.head(8)
```

```python
# Checks the datatypes of the data in titanic dataset
titanic.dtypes
```

```python
# Check if there are any missing values
titanic.isnull().values.any()
```

```python
# Drops any null values and now rechecks whether there are any nulls
titanic.dropna(inplace=True)
titanic.isnull().values.any()
```

```python
# Basic statistics of the titanic data
titanic.describe()
```

```python
# Mean on grouped data
titanic[["Sex", "Age"]].groupby("Sex").mean()
```

```python
# Findings: The average male in the titanic was older than the average female.
```

```python
# Created a line plot for the age distributiojn of the passengers
plt.plot(titanic.index, titanic['Age'], color='b', label='Age')
plt.title('Age Distribution of Passengers')
plt.xlabel('Passenger Index')
plt.ylabel('Age')
plt.legend()
plt.grid(True)
```

```python
# Created a bar graph to explain the data 
survival_rate_by_gender = titanic.groupby('Sex')['Survived'].mean()

survival_rate_by_gender.plot(kind='bar', color=['lightblue', 'pink'])
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
```

```python
# Created a histogram
plt.hist(titanic['Age'], bins=20, color='green', edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(data=titanic, x='Age', y='Fare', hue='Survived', palette='coolwarm')
plt.title('Age vs Fare by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
```
